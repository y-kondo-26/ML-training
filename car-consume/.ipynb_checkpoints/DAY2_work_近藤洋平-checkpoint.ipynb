## 線形回帰

```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error # 回帰問題における性能評価に関する関数
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

### データの読み込み

```python
df_consume = pd.read_excel("measurements2.xlsx")
```

```python
# データ確認
display(df_consume.head())
display(df_consume.describe())
print('dataframeの行数・列数の確認==>\n', df_consume.shape)
print('dataframeの各列のデータ型を確認==>\n', df_consume.dtypes)
print('欠損値の数を確認==>\n', df_consume.isnull().sum())
```

★specials, refill liters , refill gasは欠損値が多いため除外<br>
★temp_insideは平均値で埋める

```python
#除外
df_consume = df_consume.drop(['specials'], axis=1)
df_consume = df_consume.drop(['refill liters'], axis=1)
df_consume = df_consume.drop(['refill gas'], axis=1)
#欠損値０埋め
df_consume = df_consume.fillna(df_consume.mean())
```

```python
df_consume_dummy = pd.get_dummies(df_consume['gas_type'])
df_consume = pd.concat([df_consume, df_consume_dummy] , axis=1)
df_consume = df_consume.drop(['gas_type'], axis=1)
df_consume.head()
```

```python
#外気温と内気温の差を追加
df_consume['temp_diff'] = df_consume['temp_inside'] - df_consume['temp_outside']
#除外
df_consume = df_consume.drop(['temp_outside'], axis=1)
df_consume = df_consume.drop(['temp_inside'], axis=1)
```

```python
#散布図
pd.plotting.scatter_matrix(df_consume, figsize=(10,10))
plt.show()
```

```python
# 相関係数を確認
df_consume.corr()
```

```python
# 相関係数をヒートマップにして可視化
sns.heatmap(df_consume.corr())
plt.show()
```

・distanceとspeedの相関が高い　→　speedを用いる

```python
y = df_consume["consume"].values
X = df_consume[['speed', 'temp_diff',  'AC', 'rain', 'E10', 'SP98']].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0, random_state=0)
```

```python
#データの標準化
stdsc = StandardScaler()
stdsc.fit(X_train)
X_train_std = stdsc.transform(X_train)
plt.hist(X_train_std[:, 0]) #標準化後の1つ目の特徴に関するヒストグラム（学習用データ）
```

```python
X_test_std = stdsc.transform(X_test)
plt.hist(X_test_std[:, 0]) #標準化後の1つ目の特徴に関するヒストグラム（テストデータ）
```

```python
# 係数を求める
regr = LinearRegression(fit_intercept=True)
regr.fit(X_train_std, y_train)

# 重みを取り出す
w0 = regr.intercept_
w1 = regr.coef_[0]
w2 = regr.coef_[1]
w3 = regr.coef_[2]
w4 = regr.coef_[3]
w5 = regr.coef_[4]
w6 = regr.coef_[5]
```

```python
x１ = X_train_std[:, 0]
x2 = X_train_std[:, 1]
x3 = X_train_std[:, 2]
x4 = X_train_std[:, 3]
x5 = X_train_std[:, 4]
x6 = X_train_std[:, 5]

# 重みと二乗誤差の確認
y_pred_train = w0 + w1 * x1 + w2 * x2 + w3 * x3 + w4 * x4 + w5 * x5 + w6 * x6
squared_error = 0.5 * np.sum((y_train - y_pred_train) ** 2)
print('w0 = {:.3f}, w1 = {:.3f}, w2 = {:.3f}, w3 = {:.3f}, w4 = {:.3f}, w5 = {:.3f}, w6 = {:.3f}'.format(w0, w1, w2, w3, w4, w5, w6))
print('二乗誤差 = {:.3f}'.format(squared_error))
```

```python
# MSEを計算
mse_train = mean_squared_error(y_train, y_pred_train) 
print("MSE = %s"%round(mse_train,3) )  

# MAEを計算
mae_train = mean_absolute_error(y_train, y_pred_train) 
print("MAE = %s"%round(mae_train,3) )

# RMSEを計算
rmse_train = np.sqrt(mse_train)
print("RMSE = %s"%round(rmse_train, 3) )
```

DAY1の結果はMSE = 0.851、MAE = 0.618、RMSE = 0.922<br>
学習データの誤差は標準化と特徴量選択により改善した。

```python
#テストデータで予測
y_pred_test = regr.predict(X_test_std)
squared_error = 0.5 * np.sum((y_test - y_pred_test) ** 2)
print('w0 = {:.3f}, w1 = {:.3f}, w2 = {:.3f}, w3 = {:.3f}, w4 = {:.3f}, w5 = {:.3f}, w6 = {:.3f}'.format(w0, w1, w2, w3, w4, w5, w6))
print('二乗誤差 = {:.3f}'.format(squared_error))


# MSEを計算
mse_test = mean_squared_error(y_test, y_pred_test) 
print("MSE = %s"%round(mse_test,3) )  

# MAEを計算
mae_test = mean_absolute_error(y_test, y_pred_test) 
print("MAE = %s"%round(mae_test,3) )

# RMSEを計算
rmse_test = np.sqrt(mse_test)
print("RMSE = %s"%round(rmse_test, 3) )
```

テストデータのほうが誤差が大きく、過学習を起こしている
